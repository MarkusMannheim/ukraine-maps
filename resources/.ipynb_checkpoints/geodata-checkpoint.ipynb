```python
import geopandas as gpd, topojson as tp
from shapely import geometry

geodata = gpd.read_file("world.zip")
geodata = geodata[geodata.CNTRY_NAME == "Ukraine"]
geodata = geodata[["CNTRY_NAME", "geometry"]]
geodata.columns = ["name", "geometry"]
geodata.to_file("ukr.geojson", driver="GeoJSON")
geodata.plot(figsize=(15, 15))
```

```python
topo = tp.Topology(geodata, toposimplify=1e-3).topoquantize(1e4).to_json("ukr_3_4.topojson")
```
